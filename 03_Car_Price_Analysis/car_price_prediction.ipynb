```python
plt.figure(figsize=(6,4))
sns.countplot(x='Fuel_Type', data=df, palette='Set1')
plt.title('Distribution of Cars by Fuel Type')
plt.show()
```

```python
df['Transmission'].value_counts()
```

```python
plt.figure(figsize=(5,5))
df['Transmission'].value_counts().plot.pie(autopct='%1.1f%%', colors=['lightblue', 'lightgreen'], startangle=90)
plt.title('Proportion of Transmission Types')
plt.ylabel('')
plt.show()
```

```python
plt.figure(figsize=(7,4))
sns.barplot(x='Fuel_Type', y='Selling_Price', data=df, palette='Set2', ci=None)
plt.title('Average Selling Price by Fuel Type')
plt.xlabel('Fuel Type')
plt.ylabel('Selling Price (in Lakhs)')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.barplot(x='Transmission', y='Selling_Price', data=df, palette='pastel', ci=None)
plt.title('Transmission Type vs Selling Price')
plt.show()
```

```python
plt.figure(figsize=(7,4))
sns.scatterplot(x='Present_Price', y='Selling_Price', data=df, hue='Transmission', s=100)
plt.title('Original Present Price vs Selling Price')
plt.xlabel('Present Price (in Lakhs)')
plt.ylabel('Selling Price (in Lakhs)')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(7,4))
sns.scatterplot(x='Driven_kms', y='Selling_Price', data=df, color='purple', alpha=0.7)
plt.title('Kilometers Driven vs Selling Price')
plt.xlabel('Kilometers Driven')
plt.ylabel('Selling Price (in Lakhs)')
plt.show()
```

```python
plt.figure(figsize=(6,4))
numerical_df = df[['Year', 'Selling_Price', 'Present_Price', 'Driven_kms', 'Owner']]
sns.heatmap(numerical_df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix of Car Features')
plt.show()
```
